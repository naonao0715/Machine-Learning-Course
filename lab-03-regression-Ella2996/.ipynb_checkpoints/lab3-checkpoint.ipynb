```python
from pandas.plotting import scatter_matrix
attributes = ['accommodates', 'beds', 'bedrooms', 'bathrooms','guests_included']
scatter_matrix(df[attributes], figsize=(12, 8))
```

```python
df = df.dropna()
df.info()
#df_f = df[attributes].dropna()
#X = df_f[attributes[1:]]
#y = df_f['accommodates']
```

```python
X = df.drop(['accommodates'], axis=1)
Y = df.iloc[:, 2]
X.info()
Y
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(X)

X_scaled = scaler.transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.2, random_state=42)
```

```python
Y_train
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
reg = LinearRegression()
reg.fit(X_train, Y_train)

```

```python
mse = mean_squared_error(Y_test, reg.predict(X_test))
mae = mean_absolute_error(Y_test, reg.predict(X_test))
rmse = math.sqrt(mse)
print("LR: mse " + str(mse) + " mae " + str(mae) + " rmse " + str(rmse))
```

#### KNN, this is very simple algorithm, and if we have enough training example, it could work pretty well 

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, Y_train)
```

```python
mse = mean_squared_error(Y_test, knn.predict(X_test))
mae = mean_absolute_error(Y_test, knn.predict(X_test))
rmse = math.sqrt(mse)
print("KNN: mse " + str(mse) + " mae " + str(mae) + " rmse " + str(rmse))
```

### And LR performs much better than KNN in all metrics

```python

import matplotlib.pyplot as plt
```

#### GT data

```python
#### our own version of kmeans
import kmeans
import sklearn
import numpy as np
from sklearn.datasets import make_blobs

X, cluster_assignments = make_blobs(n_samples=200, centers=4, cluster_std=0.60, random_state=0)

k = 4

color = ['red', 'green', 'blue', 'yellow', 'black']
for i in range(k):
    plt.scatter(X[cluster_assignments == i , 0] , X[cluster_assignments == i , 1] , label = i)
    print("cluster " + str(i) + " has " + str(np.sum(cluster_assignments == i)) + " samples")
plt.legend()
plt.show()
```

#### normal KMeans

```python
kmean = kmeans.KMeans(k = 4)
clusters_predicted, clusters_center = kmean.fit(X)

for i in range(k):
    plt.scatter(X[clusters_predicted == i , 0] , X[clusters_predicted == i , 1] , label = i)
    print("cluster " + str(i) + " has " + str(np.sum(clusters_predicted == i)) + " samples")
plt.scatter(clusters_center[:,0] , clusters_center[:,1] , s = 8, color = 'k')
plt.legend()
plt.show()

```

#### extended kmeans

```python
clusters_predicted2, clusters_center2 = kmean.fit_extended(X, True)

for i in range(k):
    plt.scatter(X[clusters_predicted2 == i , 0] , X[clusters_predicted2 == i , 1] , label = i)
    print("cluster " + str(i) + " has " + str(np.sum(clusters_predicted2 == i)) + " samples")
plt.scatter(clusters_center2[:,0] , clusters_center2[:,1] , s = 8, color = 'k')
plt.legend()
plt.show()


```

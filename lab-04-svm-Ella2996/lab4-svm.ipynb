```python
X = df.iloc[:, [1,3,5,8,9]]
Y = df['label']
X
```

```python
from sklearn.model_selection import train_test_split

from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(X)

X_scaled = scaler.transform(X)

X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.3, random_state=42)
```

## Training  data ratio: ~70%
## Testing data ratio: ~30%

```python
print("train ratio: " + str(len(X_train) / len(X)))
print("test ratio: " + str(len(X_test) / len(X)))
```

```python
from sklearn.svm import SVC
import numpy as np
clf_linear = SVC(kernel = 'linear')
clf_linear.fit(X_train, Y_train)
y_predicted_linear = clf_linear.predict(X_test)

err_linear = np.mean(y_predicted_linear != Y_test)
err_linear
```

```python
clf_poly = SVC(kernel = 'poly')

clf_poly.fit(X_train, Y_train)
y_predicted_poly = clf_poly.predict(X_test)

err_poly = np.mean(y_predicted_poly != Y_test)
err_poly
```

```python
clf_rbf = SVC(kernel = 'rbf')

clf_rbf.fit(X_train, Y_train)
y_predicted_rbf = clf_rbf.predict(X_test)

err_rbf = np.mean(y_predicted_rbf != Y_test)
err_rbf
```

### RBF gives the best performance 

```python
regularizations = pd.DataFrame(data=[], index=range(1, 11), columns=['ER'])

for c in range(1, 11):
    print(c)
    clf = SVC(C = c, kernel = 'rbf')
    clf.fit(X_train, Y_train)
    y_predicted = clf.predict(X_test)
    regularizations.loc[c] = np.mean(y_predicted != Y_test)
regularizations
```

####  The best regularization is when C =  8, 9, 10, we need a large C value, mostly because we are overfitting to the training data, so we need a large C value to reduce the overfittting

```python
ax = plt.gca()

regularizations.plot(kind='line', y='ER', ax=ax)
plt.show()
```

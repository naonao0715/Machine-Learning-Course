```python
import pandas as pd

df = pd.read_csv("lab05_dataset.csv")

#df  = df.sample(n = 100000)
#df
```

```python
df.describe()
```

```python
X = df.iloc[:, :-1]
Y = df.iloc[:, -1]


from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

## No preprocessing is required as it already achieves > 90% accuracy with baseline decision tree model. 

```python
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
clf = tree.DecisionTreeClassifier()
clf.fit(X_train,Y_train)
```

```python
hyp_test = clf.predict(X_test)
acc = accuracy_score(Y_test, hyp_test)
print("baseline accuracy: " + str(acc))
```

```python
clf.feature_importances_
```

### 11 features are needed to achieve +-2% accuracy. 

```python
num_feature = 11
idxes = clf.feature_importances_.argsort()[-num_feature:][::-1]

X_train_s = X_train.iloc[:, idxes]
X_test_s = X_test.iloc[:, idxes]

clf_s = tree.DecisionTreeClassifier()
clf_s.fit(X_train_s,Y_train)

hyp_test = clf_s.predict(X_test_s)
acc = accuracy_score(Y_test, hyp_test)
acc
```

### Based on the confusion matrix and the classification report, it's showed that Random Forest and Bagging give the best accuracy of around 95%. And AdaBoost preforms the worse. It's interesting that all methods give relatively low recall for class 5. 

## ExtraTree

```python
from sklearn.tree import ExtraTreeClassifier
extra_tree = ExtraTreeClassifier()
extra_tree.fit(X_train_s,Y_train)

hyp_test = extra_tree.predict(X_test_s)
print(accuracy_score(Y_test, hyp_test))
print(confusion_matrix(Y_test, hyp_test))
print(classification_report(Y_test, hyp_test))
```

## random forest

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier()
random_forest.fit(X_train_s,Y_train)

hyp_test = random_forest.predict(X_test_s)
print(accuracy_score(Y_test, hyp_test))
print(confusion_matrix(Y_test, hyp_test))
print(classification_report(Y_test, hyp_test))
```

## XGBoost

```python
import xgboost as xgb
xfbo =xgb.XGBClassifier()
xfbo.fit(X_train_s,Y_train)

hyp_test = xfbo.predict(X_test_s)
print(accuracy_score(Y_test, hyp_test))
print(confusion_matrix(Y_test, hyp_test))
print(classification_report(Y_test, hyp_test))
```

## AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier

ada = AdaBoostClassifier()
ada.fit(X_train_s,Y_train)
hyp_test = ada.predict(X_test_s)
print(accuracy_score(Y_test, hyp_test))
print(confusion_matrix(Y_test, hyp_test))
print(classification_report(Y_test, hyp_test))
```

## Bagging

```python
from sklearn.ensemble import BaggingClassifier
bag = BaggingClassifier()
bag.fit(X_train_s,Y_train)
hyp_test = bag.predict(X_test_s)
print(accuracy_score(Y_test, hyp_test))
print(confusion_matrix(Y_test, hyp_test))
print(classification_report(Y_test, hyp_test))
```

```python
import pandas as pd
## load the dataset
df = pd.read_csv('credit_fraud_class.csv')

### there is no null value in the dataset
df.info()
```

```python
df.dropna() 
```

```python
## check distribution
import matplotlib.pyplot as plt
df.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
X = df.iloc[:, :29]
Y = df.iloc[:, 29]
num_feature = X.shape[1]
```

```python
### compute feature correlation 
corr = df.corr()
abs(corr['Class']).nlargest(2)
### as you can see the maximum correlating value is -0.646226 from V14
```

```python
## visualize in heat map 
import seaborn as sns; sns.set_theme()
ax = sns.heatmap(corr)
# will use column 3 Number of Proposed Stories as input and column 2 Number of Existing Stories as output 
```

```python
plt.subplots(figsize=(18,10 ))
for i in range(num_feature):   
    plt.subplot(6, 5, i + 1)
    plt.scatter(X.iloc[:, i], Y[:])

plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = num_feature)
pca.fit(X_scaled)
```

```python
import numpy as np
cvar =np.cumsum(pca.explained_variance_ratio_)
plt.plot(cvar)
plt.xlabel('no. of principal components')
plt.ylabel('cumulative variance')
```

```python
cvar
```

```python
idx_90 = np.argmax(cvar > 0.9)
idx_90
```

```python
pca_90 = PCA(n_components = idx_90 + 1)
X_pca = pca_90.fit_transform(X_scaled)
```

```python
df2 = pd.DataFrame(X_pca)
df2
```

```python
df2 = pd.concat([df2, Y], axis = 1)
df2
```

```python
### with PCA, the maximum correlation increases to 0.758 and it's from feature 0. 
### Seems like the remaining featurs are not that related to the prediction actually 
corr2 = df2.corr()
abs(corr['Class']).nlargest(2)
```
